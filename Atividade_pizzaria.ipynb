```python

# Loop principal para permitir múltiplos pedidos
while True:
  # Exibe o menu da pizzaria
  print("-----------Bem-vindo à Pizzaria-----------------")
  print("")
  print("------------------Cardápio----------------------")
  print("")
  print(f"{'Tamanho'} | {'Pizza Salgada(PS)'} | {'Pizza Doce(PD)'}")
  print(f"{'P'} | {'R$ 30,00'} | {'R$ 34,00'}")
  print(f"{'M'} | {'R$ 45,00'} | {'R$ 48,00'}")
  print(f"{'G'} | {'R$ 60,00'} | {'R$ 66,00'}")
  print("")
  print("-------------Refrigerante 2L  R$ 15,00----------")
  print("")
  print("----AVISO: NÃO REALIZAMOS ENTREGAS ACIMA DE 25KM----")
  print("")

  # Inicializa as variáveis para um novo pedido
  total_pedido = 0.0
  num_pizzas = 0
  num_refrigerantes = 0

  # Dicionários com os preços das pizzas e refrigerante
  pizza_salgada_prices = {'P': 30.00, 'M': 45.00, 'G': 60.00}
  pizza_doce_prices = {'P': 34.00, 'M': 48.00, 'G': 66.00}
  refrigerante_price = 15.00

  print("\nIniciando um novo pedido:")
  # Loop para adicionar itens ao pedido
  while True:
    item_desejado = input("Qual item você deseja (Pizza/Refrigerante/Finalizar): ").lower()

    # Verifica se o usuário quer finalizar o pedido
    if item_desejado == 'finalizar':
      break
    # Processa a adição de uma pizza
    elif item_desejado == 'pizza':
      # Limite de 5 pizzas por pedido
      if num_pizzas < 5:
        sabor_pizza = input("Qual sabor de pizza você deseja (PS/PD): ").upper()
        tamanho_pizza = input("Qual tamanho da pizza você deseja (P/M/G): ").upper()

        # Adiciona a pizza se o sabor e tamanho forem válidos
        if sabor_pizza == 'PS' and tamanho_pizza in pizza_salgada_prices:
          pizza_price = pizza_salgada_prices[tamanho_pizza]
          pizza_type_name = 'Salgada'
          total_pedido += pizza_price
          num_pizzas += 1
          print(f"Pizza {pizza_type_name} {tamanho_pizza} adicionada. Preço: R$ {pizza_price:.2f}")
          print(f"Total parcial do pedido: R$ {total_pedido:.2f}")
        elif sabor_pizza == 'PD' and tamanho_pizza in pizza_doce_prices:
          pizza_price = pizza_doce_prices[tamanho_pizza]
          pizza_type_name = 'Doce'
          total_pedido += pizza_price
          num_pizzas += 1
          print(f"Pizza {pizza_type_name} {tamanho_pizza} adicionada. Preço: R$ {pizza_price:.2f}")
          print(f"Total parcial do pedido: R$ {total_pedido:.2f}")
        else:
          print("Opção de pizza inválida. Por favor, escolha PS ou PD para o sabor e P, M ou G para o tamanho.")
      else:
        print("Limite de 5 pizzas atingido!")
    # Processa a adição de um refrigerante
    elif item_desejado == 'refrigerante':
      # Limite de 2 refrigerantes por pedido
      if num_refrigerantes < 2:
        total_pedido += refrigerante_price
        print(f"Refrigerante adicionado. Preço: R$ {refrigerante_price:.2f}")
        num_refrigerantes += 1
        print(f"Total parcial do pedido: R$ {total_pedido:.2f}")
      else:
        print("Limite de 2 refrigerantes atingido!")
    # Mensagem para item inválido
    else:
      print("Item inválido. Por favor, escolha 'Pizza', 'Refrigerante' ou 'Finalizar'.")

  print(f"\nSeu total de pedido é: R$ {total_pedido:.2f}")

  # Calcula a taxa de entrega baseada na distância
  try:
    distancia = float(input("Qual a distância até a sua residência (em km)?: "))
    if distancia > 25:
      print("Não realizamos entregas acima de 25km.")
    elif 5 < distancia <= 10:
      total_pedido += 10.00
      print("Adicionado R$ 10.00 de taxa de entrega.")
    elif distancia > 10:
      taxa_adicional = 10.00 + (distancia - 10) * 1.10
      total_pedido += taxa_adicional
      print(f"Adicionado R$ {taxa_adicional:.2f} de taxa de entrega.")
    else: # distancia <= 5
      print("Taxa de entrega gratuita para distâncias até 5km.")
    print(f"Total do pedido com entrega: R$ {total_pedido:.2f}")
  except ValueError:
    print("Distância inválida. Por favor, insira um número.")

  # Pergunta se o usuário deseja fazer outro pedido
  continuar = input("Deseja fazer outro pedido? (sim/não): ").lower()
  if continuar != 'sim':
    print("Obrigado por pedir na nossa Pizzaria!")
    break
```
