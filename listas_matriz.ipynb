```python
#criando lista somente com nomes com a letra A
nomes = ["Bruno", "Maria", "Marcos", "Pedro", "Claudio"]
nomes_A = []
for x in nomes:
  if "a" in x.lower():
    nomes_A.append(x)
print(nomes_A)
```

```python
#lista de nomes com 2 S
nomes = ["Clarissa", "Vanessa", "Jéssica", "Clarice", "Susana"]
nomes_SS = []
for x in nomes:
  if x.lower().count("s") == 2:
    nomes_SS.append(x)
print(nomes_SS)
```

```python
#lista de nomes que terminam com O
nomes = ["Clarissa", "Vanessa", "Jéssica", "Clarice", "Susana", "Bruno", "Pedro"]
nomes_O = []
for x in nomes:
  if x.lower()[-1] == "o":
    nomes_O.append(x)

nomes_O.sort(reverse=True)
print(nomes_O)
```

```python
#tabuada do 1 ao 10
for i in range(1,11):
  for j in range(1,11):
    print(f"{i} x {j} = {i*j}")
  print()
```
