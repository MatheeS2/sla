```python
pessoas = []
pessoas.append({"nome":"Bruno", "sexo": "M"})
pessoas.append({"nome":"Maria", "sexo": "F"})
pessoas.append({"nome":"João", "sexo": "M"})
pessoas.append({"nome":"Ana", "sexo": "F"})


for pessoa in pessoas:
  s = pessoa["sexo"]
  n = pessoa["nome"]
  print(f"Bem-vindo Sr. {n}" if s == "M" else f"Bem-vinda Sra. {n}")
```

```python
def somaVaslores():
  val1 = float(input("Digite o primeiro valor: "))
  val2 = float(input("Digite o segundo valor: "))
  print(f"{val1} + {val2} = {val1 + val2}")

somaVaslores()
```

```python
def somaValores(a, b):
  print(f"{a} + {b} = {a + b}")

somaValores(1, 3)
somaValores(45.8, 68)
somaValores(777, -666)
```

```python
def tabuada(n):
  for i in range(1, 11):
    print(f"{n} x {i} = {n * i}")
  print("Fim da tabuada")

num = int(input("Qual tabuada deseja saber?: "))
tabuada(num)
```

```python
def hipotenusa(cA, cB):
  return (cA**2 + cB**2)**(1/2)

a = float(input("Digite o valor do cateto adjacente: "))
b = float(input("Digite o valor do cateto oposto: "))
hipo =  hipotenusa(a,b)
print(f"A hipotenusa é {hipo}")
```

```python
import math
while True:
  def fatorial(A):
    if A < 0:
      print("Não existe fatorial de número negativo")

    return  math.factorial(A)

  a = int(input("Digite o valor que deseja saber o fatorial: "))
  fat =  fatorial(a)

  print(f"A fatorial é {fat}")
  break
```

```python
def bhaskara(a, b, c):
    delta = b**2 - 4*a*c

    if delta < 0:
        return "Não existem raízes reais"

    x1 = (-b + delta**0.5) / (2*a)
    x2 = (-b - delta**0.5) / (2*a)

    return x1, x2

a = float(input("Digite o valor A: "))
b = float(input("Digite o valor B: "))
c = float(input("Digite o valor C: "))

resultado = bhaskara(a, b, c)
print(f"O resultado é: {resultado}")
```
