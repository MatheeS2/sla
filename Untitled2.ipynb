```python
soma = 0
maior = float('-inf')
menor = float('inf')
while True:
  valor = float(input("Qual valor você deseja que seja somado(valor com números entre 1 e 100): "))
  if valor == 0:
    break
  if valor < 0:
    print("Valor inválido")
    continue
  if valor > 100:
    print("Valor inválido")
    continue
  soma += valor
  if valor > maior:
    maior = valor
  if valor < menor:
    menor = valor
print(f"A soma dos valores deu {soma}")
if maior == float('-inf'):
  print("Nenhum valor válido foi informado.")
else:
  print(f"O maior valor informado foi {maior}")
  print(f"O menor valor informado foi {menor}")
```
