```python
while True:
  print("Menu de opções: ")
  print("1 - Cadastrar")
  print("2 - Consultar")
  print("3 - Sair")
  opc = input(">>")
  match opc.lower():
    case "1":
      print("Cadastrando")
    case "2":
      print("Consultandov")
    case "3":
      print("Até logo")
      break
    case _:
      print("Opção inválida, tente novamente")
```

```python
valor1= 0
valor2= 0
while True:
  print("Menu de opções: ")
  print("1 - +")
  print("2 - -")
  print("3 - *")
  print("3 - /")
  valor1 = float(input("Qual o primeiro valor?: "))
  valor2 = float(input("Qual o segundo valor?: "))
  print("Qual operação deseja realizar?: ")
  opc = input(">>")
  match opc.lower():
    case "1":
      print(valor1 + valor2)
      break
    case "2":
      print(valor1 - valor2)
      break
    case "3":
      print(valor1 * valor2)
      break
    case "4":
      print(valor1 / valor2)
      if valor2 or valor1 == 0:
        print("Não é possível dividir por zero")
      break
    case _:
      print("Opção inválida, tente novamente")
```
